```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def step_function(x):
    y =  x > 0
    return y.astype(np.int)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.title('Step function')
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.title('Sigmoid function')
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
def ReLU(x):
    return np.maximum(0, x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = ReLU(x)
plt.plot(x, y)
plt.title('ReLU function')
plt.ylim(-1.0, 6.0)
plt.show()
```
